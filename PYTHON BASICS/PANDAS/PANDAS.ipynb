```python
#reading excel

def convert_people_cell(cell):
    if cell =="n.a.":
        return "sam walten"
    return cell

def convert_eps_cell(cell):
    if cell =="not available":
        return None
    return cell

#using converters

df = pd.read_excel('stock_data.xlsx','Sheet1',converters={
    'people':convert_people_cell,
    'eps': convert_eps_cell
})
df
```

```python
df.to_excel('new.xlsx',sheet_name='Stocks',startrow=1,startcol=2)
```

#### Handling Missing Data : fillna,dropna , interpolate

```python
# parse data used to convert str date to timestamp
df =pd.read_csv('weather_data5.csv',parse_dates=['day'])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
```

```python
df
```

```python
# fillna
df.fillna(0)
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'noevent'
})
new_df
```

```python
#used to place previous days values : forwardfill
new_df =df.fillna(method='ffill')
new_df
```

```python

#used to place previous days values : backfill
new_df =df.fillna(method='bfill')
new_df
```

```python

#used to place previous days values : side column
new_df =df.fillna(method='bfill',axis='columns')
new_df
```

```python
#used to place previous days values : forwardfill with one forward
new_df =df.fillna(method='ffill',limit=1)
new_df
```

```python
#interpolate
new_df = df.interpolate()
new_df
```

```python
#interpolate
new_df = df.interpolate(method='time')
new_df
```

```python
#dropna
new_df = df.dropna()
new_df
```

```python
#drop row that has all na
new_df = df.dropna(how='all')
new_df
```

```python
#drop row that has valid thresh value
new_df = df.dropna(thresh=2)
new_df
```

```python
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

#### Crosstab

```python
pd.crosstab(df.Nationality,df.Handedness)
```

```python
pd.crosstab(df.Sex,df.Handedness)
```

```python
pd.crosstab(df.Nationality,df.Handedness,margins=True)
```

```python
pd.crosstab(df.Sex,[df.Handedness,df.Nationality],margins=True)
```

```python
pd.crosstab([df.Sex,df.Nationality],[df.Handedness],margins=True)
```

```python
pd.crosstab(df.Sex,df.Handedness,normalize='index')
```
