```python
# what if using other calendar like india
#Build out own calendar
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday
class myBirthDayCalendar(AbstractHolidayCalendar):
    rules=[
        Holiday('Praveen birth day',month=4, day=15,observance=nearest_workday)#nearest workday to mark leave near leave workdays
    ]
myc = CustomBusinessDay(calendar=myBirthDayCalendar())
myc
```
